```python
!pip install transformers torch flask-ngrok

from transformers import pipeline
import torch

# Chatbot pipeline (you can customize or upgrade to more advanced models)
chatbot = pipeline("text-generation", model="gpt2")

# Sentiment analyzer
sentiment = pipeline("sentiment-analysis")

def get_bot_response(user_input):
    chat_reply = chatbot(user_input, max_length=50, do_sample=True)[0]['generated_text']
    sentiment_label = sentiment(user_input)[0]['label']
    return chat_reply.strip(), sentiment_label

# Test
user_message = "I'm not happy with the garbage collection service."
reply, mood = get_bot_response(user_message)
print(f"Bot Reply: {reply}\nSentiment: {mood}")
```
